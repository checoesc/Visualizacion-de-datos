**Creaci칩n de Scatterplot por categor칤as y con variables condicionantes**

Para poder crear este tipo de gr치fico, utilizaremos un dataset con informaci칩n sobre ataques cibern칠ticos que encontraremos en el archivo Ejemplo_02_03_Cybersecurity_Dataset donde encontraremos las columnas:

**Aplica los siguientes pasos:**

1. Creaci칩n del scatterplot: Este c칩digo genera un scatterplot b치sico que muestra la relaci칩n entre la severidad de los incidentes y el tiempo de respuesta.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
cybersecurity_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/EUREKA/DATASET/Ejemplo_02_03_Cybersecurity_Dataset.csv') # Cambia la ruta del archivo si es necesario

# Configurar el estilo de Seaborn
sns.set(style="darkgrid")

# Crear el Scatterplot sin dividir por categor칤as
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cybersecurity_data, x='severity', y='response_time', color='b', s=100)

# A침adir t칤tulo y etiquetas a los ejes
plt.title('Scatterplot de Severidad vs Tiempo de respuesta', fontsize=16)
plt.xlabel('Severidad del incidente', fontsize=14)
plt.ylabel('Tiempo de respuesta (horas)', fontsize=14)

# Mostrar la gr치fica
plt.show()
```

sns.scatterplot(data=cybersecurity_data, x='severity', y='response_time', color='b', s=100):

Se crea un scatterplot simple, donde todos los puntos son del mismo color (color='b' para azul).

x='severity' y y='response_time' especifican los ejes del gr치fico.
s=100 ajusta el tama침o de los puntos.


 2. Scatterplot por categor칤a: Para analizar c칩mo var칤an la severidad y el tiempo de respuesta seg칰n las categor칤as de ataque m치s frecuentes, un scatterplot por categor칤a es ideal.

```python
# Verificar la frecuencia de cada tipo de ataque para identificar los dos m치s comunes
attack_type_counts = cybersecurity_data['attack_type'].value_counts()
top_two_attack_types = attack_type_counts.index[:2]

# Filtrar el dataset para incluir solo los dos tipos de ataques m치s comunes
filtered_data = cybersecurity_data[cybersecurity_data['attack_type'].isin(top_two_attack_types)]

# Crear el Scatterplot para los dos tipos de ataques m치s comunes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_data, x='severity', y='response_time', hue='attack_type', palette='muted', s=100)

# A침adir t칤tulo y etiquetas a los ejes
plt.title('Severidad vs Tiempo de respuesta (2 ataques m치s comunes)', fontsize=16)
plt.xlabel('Severidad del incidente', fontsize=14)
plt.ylabel('Tiempo de respuesta (horas)', fontsize=14)

# Mostrar la gr치fica
plt.show()
```

**Creaci칩n de Treemaps**

Un Treemap es una t칠cnica visual que muestra datos jer치rquicos o categ칩ricos mediante rect치ngulos de diferentes tama침os y colores. Cada rect치ngulo representa una categor칤a, y su tama침o y color reflejan la proporci칩n o el valor de los datos, facilitando la comparaci칩n de m칰ltiples categor칤as en una sola visualizaci칩n.

Plotly Express: Biblioteca interactiva en Python que permite crear gr치ficos din치micos y personalizables. Es ideal para presentaciones, dashboards, y explorar datos en profundidad con interactividad.

Squarify: Biblioteca especializada en crear treemaps est치ticos. Es adecuada para generar gr치ficos r치pidos y simples cuando no se requiere interactividad, manteniendo la eficiencia.

**Treemap con Plotly Express**

Los pasos que haremos para iniciar son los que ya conoces: Instalaremos las bibliotecas, en este caso Plotly y Pandas, cargaremos el dataset (seguiremos usando el de ciberseguridad) y despu칠s crearemos el Treemap con Plotly Express:


```python
!pip install plotly
```

```python
import pandas as pd
import plotly.express as px

# Cargar el dataset
cybersecurity_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/EUREKA/DATASET/Ejemplo_02_03_Cybersecurity_Dataset.csv') # Cambia la ruta del archivo si es necesario

# Contar la cantidad de incidentes por tipo de ataque
incident_counts = cybersecurity_data['attack_type'].value_counts().reset_index()
incident_counts.columns = ['attack_type', 'count']

# Crear el Treemap basado en el conteo de incidentes
fig = px.treemap(incident_counts,
                 path=['attack_type'],
                 values='count',
                 color='count',
                 color_continuous_scale='Oranges',
                 title='Treemap de Tipos de Ataque en Ciberseguridad')

# Mostrar el Treemap
fig.show()
```

쯈ue estamos haciendo?

Agrupar los datos:

value_counts(): Cuenta la frecuencia de cada tipo de ataque en el dataset.

reset_index(): Resetea el 칤ndice para que los datos puedan ser utilizados f치cilmente en el treemap.

px.treemap:

path=['attack_type']: Especifica la jerarqu칤a de los datos, en este caso, estamos creando el treemap basado en el tipo de ataque (attack_type).

values='count': Usamos un conteo de incidentes por tipo de ataque, lo que asegura que estamos pasando valores num칠ricos y agregables.

color='count': Asigna un color a cada bloque basado en el conteo. Puedes cambiar esta variable por otra que te interese resaltar.

color_continuous_scale='Oranges': Define la paleta de colores utilizada. Puedes reemplazar 'Oranges' por otras paletas como 'Reds', 'Purples', 'Oranges', 'Greens', 'Viridis', 'Cividis', 'Plasma', etc.

Interactividad:

El treemap generado por Plotly es interactivo. Puedes pasar el cursor sobre los bloques para ver detalles adicionales y hacer clic en los bloques para desglosar la informaci칩n

alpha=.8: Ajusta la transparencia de los colores.

**Creaci칩n de gr치fica de barras apiladas**

Una gr치fica de barras apiladas es una visualizaci칩n que muestra la composici칩n de diferentes categor칤as dentro de un conjunto de datos. Cada barra se divide en segmentos apilados, donde cada segmento representa la contribuci칩n de una categor칤a al total, facilitando la comparaci칩n de categor칤as tanto dentro de una barra como entre distintas barras mediante el uso de colores.

Esto podemos realizarlo de la siguiente manera:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
cybersecurity_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/EUREKA/DATASET/Ejemplo_02_03_Cybersecurity_Dataset.csv') # Cambia la ruta del archivo si es necesario

# Crear una tabla cruzada que muestre el n칰mero de incidentes por severidad y tipo de ataque
crosstab_data = pd.crosstab(cybersecurity_data['severity'], cybersecurity_data['attack_type'])

# Crear la gr치fica de barras apiladas
crosstab_data.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# A침adir t칤tulo y etiquetas a los ejes
plt.title('Gr치fica de Barras apiladas de tipos de ataque por severidad', fontsize=14)
plt.xlabel('Severidad del incidente', fontsize=12)
plt.ylabel('N칰mero de incidentes', fontsize=12)
plt.legend(title='attack_type', bbox_to_anchor=(1, 1), loc='upper left')

# Mostrar la gr치fica
plt.show()
```

쯈ue estamos haciendo?

**Agrupar los datos:**

pd.crosstab: Este m칠todo crea una tabla cruzada que cuenta la cantidad de incidentes para cada combinaci칩n de severidad y tipo de ataque.
Crear la gr치fica de barras apiladas:

kind='bar', stacked=True: Genera una gr치fica de barras apiladas. Cada barra representa una severidad y est치 dividida en segmentos que representan los diferentes tipos de ataque.

colormap='viridis': Aplica una paleta de colores a las barras apiladas. Puedes cambiar viridis por otra paleta de colores si lo prefieres.

**Personalizaci칩n:**

plt.title: A침ade un t칤tulo descriptivo a la gr치fica.
plt.xlabel y plt.ylabel: Etiquetas para los ejes x e y.
bbox_to_anchor=(1, 1): Ajusta la posici칩n de la leyenda fuera del 치rea del gr치fico. 1 mueve la leyenda hacia la derecha del gr치fico, mientras que 1 la alinea con la parte superior del gr치fico.
loc='upper left': Establece la posici칩n de la leyenda en la parte superior izquierda del cuadro delimitador especificado por bbox_to_anchor


**Interpretaci칩n de gr치fica de barras apiladas**
